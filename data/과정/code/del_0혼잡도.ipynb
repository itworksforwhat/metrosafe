```python
import pandas as pd

# 1. íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
file_path = '/Users/mac/Desktop/myPycode/ì²­ë…„ AI ìŠ¤ì¿¨ í”„ë¡œì íŠ¸/ì„œìš¸êµí†µê³µì‚¬_ì§€í•˜ì² í˜¼ì¡ë„ì •ë³´_20231231.csv'  # â† ì—¬ê¸°ì— ì‹¤ì œ íŒŒì¼ê²½ë¡œ ì…ë ¥
encodings = ['utf-8-sig', 'cp949', 'euc-kr']
for enc in encodings:
    try:
        df = pd.read_csv(file_path, encoding=enc)
        print(f"âœ… Loaded with encoding: {enc}")
        break
    except Exception as e:
        continue

# 2. ì‹œê°„ëŒ€ í˜¼ì¡ë„ ì—´ë§Œ ì„ íƒ (5ì‹œ30ë¶„ ~ 00ì‹œ30ë¶„ ì‚¬ì´)
# ë¬¸ìì—´ì´ ìˆ«ìì²˜ëŸ¼ ë³´ì´ëŠ” ê²½ìš°ë„ ìˆìœ¼ë¯€ë¡œ floatë¡œ ë³€í™˜ ì‹œë„
time_columns = df.columns[6:]  # ì•ì˜ ë©”íƒ€ë°ì´í„° 6ê°œ ì—´ ì œì™¸í•œ ëª¨ë“  ì—´

# 3. ê° ì—´ì„ ìˆ«ìë¡œ ë³€í™˜ (ì—ëŸ¬ ë¬´ì‹œí•˜ê³  ìˆ«ì ì•„ë‹Œ ê±´ NaN ì²˜ë¦¬)
df[time_columns] = df[time_columns].apply(pd.to_numeric, errors='coerce')

# 4. í˜¼ì¡ë„ í•©ì´ 0 (ë˜ëŠ” NaN ì œì™¸í•˜ê³  0ì¸ ê²½ìš°)ì¸ í–‰ ì°¾ê¸°
zero_congestion_mask = (df[time_columns].fillna(0).sum(axis=1) == 0)

# 5. ê²°ê³¼ ì¶œë ¥
print(f"ğŸš¨ í˜¼ì¡ë„ ì´í•©ì´ 0ì¸ í–‰ ê°œìˆ˜: {zero_congestion_mask.sum()}ê°œ")

# 6. 0ì¸ í–‰ ì œê±°
df_cleaned = df[~zero_congestion_mask]

# 7. ì €ì¥
df_cleaned.to_csv("cleaned_í˜¼ì¡ë„_0ì œê±°.csv", index=False, encoding='utf-8-sig')
print("âœ… ì œê±° ë° ì €ì¥ ì™„ë£Œ â†’ cleaned_í˜¼ì¡ë„_0ì œê±°.csv")
```
