```python
import pandas as pd

file_path = "ì—¬ê¸°ì— ë°ì´í„° íŒŒì¼ê²½ë¡œ ë³µì‚¬"

# ì¸ì½”ë”© ì‹œë„
encodings = ['utf-8-sig', 'euc-kr', 'cp949']
for enc in encodings:
    try:
        df = pd.read_csv(file_path, encoding=enc)
        print(f"âœ… Loaded with encoding: {enc}")
        break
    except:
        continue

# ì—´ ì´ë¦„ ì •ë¦¬ (í˜¹ì‹œ ëª¨ë¥¼ ê³µë°±/ìŠ¬ë˜ì‹œ ì œê±°)
df.columns = df.columns.str.replace('/', '', regex=False).str.replace(' ', '').str.strip()

# ì‹œê°„ëŒ€ ì—´ ì¶”ì¶œ (ì•ì—ì„œ 6ì—´ ì´í›„ê°€ ì‹œê°„ëŒ€ì¼ ê²½ìš°)
time_columns = df.columns[6:]

# ì‹œê°„ëŒ€ë³„ í•©ì´ 0ì¸ í–‰ í•„í„°ë§ (ìŠ¹ì°¨/í•˜ì°¨ ìƒê´€ ì—†ìŒ)
mask = (df[time_columns].sum(axis=1) == 0)

# ğŸ‘‰ ëª‡ í–‰ì¸ì§€ ì¶œë ¥
print(f"âœ… ì‹œê°„ëŒ€ í•©ê³„ê°€ 0ì¸ ì „ì²´ í–‰ ê°œìˆ˜: {mask.sum()}ê°œ")

# ğŸ‘‰ í•´ë‹¹ í–‰ ì œê±°
df_cleaned = df[~mask]

# ğŸ‘‰ ê²°ê³¼ ì €ì¥
df_cleaned.to_csv("cleaned_ì‹œê°„ëŒ€í•©0ì œê±°.csv", index=False, encoding='utf-8-sig')
print("ğŸ‰ 'cleaned_ì‹œê°„ëŒ€í•©0ì œê±°.csv' ì €ì¥ ì™„ë£Œ")
```
