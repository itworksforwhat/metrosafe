```python
import pandas as pd

file_path = "../samename_data/ì„œìš¸êµí†µê³µì‚¬_1~8í˜¸ì„ ì¼ë³„ì—­ë³„ì‹œê°„ëŒ€ë³„ìŠ¹í•˜ì°¨ì¸ì›_2015ë…„_íœ´ì¼ìˆ˜ì •ë³¸_ì´ë¦„í†µì¼í™”.csv"

# ì¸ì½”ë”© ì‹œë„
df = pd.read_csv(file_path, encoding='euc-kr')

# ì—´ ì´ë¦„ ì •ë¦¬ (í˜¹ì‹œ ëª¨ë¥¼ ê³µë°±/ìŠ¬ë˜ì‹œ ì œê±°)
df.columns = df.columns.str.replace('/', '', regex=False).str.replace(' ', '').str.strip()

# ì‹œê°„ëŒ€ ì—´ ì¶”ì¶œ (ì•ì—ì„œ 6ì—´ ì´í›„ê°€ ì‹œê°„ëŒ€ì¼ ê²½ìš°)
time_columns = df.columns[7:]
print(time_columns)

# ì‹œê°„ëŒ€ë³„ í•©ì´ 0ì¸ í–‰ í•„í„°ë§ (ìŠ¹ì°¨/í•˜ì°¨ ìƒê´€ ì—†ìŒ)
mask = (df[time_columns].sum(axis=1) == 0)

# ğŸ‘‰ ëª‡ í–‰ì¸ì§€ ì¶œë ¥
print(f"âœ… ì‹œê°„ëŒ€ í•©ê³„ê°€ 0ì¸ ì „ì²´ í–‰ ê°œìˆ˜: {mask.sum()}ê°œ")

# ğŸ‘‰ í•´ë‹¹ í–‰ ì œê±°
df_cleaned = df[~mask]

# ğŸ‘‰ ê²°ê³¼ ì €ì¥
df_cleaned.to_csv("../no_null_data/ì„œìš¸êµí†µê³µì‚¬_ë°ì´í„°_íœ´ì¼ìˆ˜ì •ë³¸_2015ë…„_nullì œê±°.csv", index=False, encoding='cp949')
print("ğŸ‰ 'cleaned_ì‹œê°„ëŒ€í•©0ì œê±°.csv' ì €ì¥ ì™„ë£Œ")
```
