```python
import os
import pandas as pd

# 1. ê²½ë¡œ ì„¤ì •
input_folder = '../2.í˜¼ì¡ë„/1_excel_data'
output_folder = '../2.í˜¼ì¡ë„/4_same_weekend_data'
os.makedirs(output_folder, exist_ok=True)

# 2. í´ë” ë‚´ ëª¨ë“  íŒŒì¼ ì²˜ë¦¬
for filename in os.listdir(input_folder):
    if not filename.endswith('.csv'):
        continue

    file_path = os.path.join(input_folder, filename)
    print(f"ğŸ“‚ ì²˜ë¦¬ ì¤‘: {filename}")

    try:
        df = pd.read_csv(file_path, encoding='euc-kr')
    except Exception as e:
        print(f"âŒ {filename} ì½ê¸° ì‹¤íŒ¨: {e}")
        continue


    # êµ¬ë¶„(ìƒì„ /í•˜ì„ ) ê³µë°± ì œê±°
    df["êµ¬ë¶„"] = df["êµ¬ë¶„"].str.strip()

    # ì£¼ë§ í‰ê·  ê³„ì‚°
    weekend = df[df["í‰ì¼ì£¼ë§"].isin(["í† ìš”ì¼", "ì¼ìš”ì¼"])]
    weekend_avg = weekend.groupby(["í˜¸ì„ ", "ì—­ë²ˆí˜¸", "ì—­ëª…", "êµ¬ë¶„"], as_index=False).mean(numeric_only=True)
    weekend_avg["í‰ì¼ì£¼ë§"] = "ì£¼ë§"

    # ì—´ ìˆœì„œ ì •ë¦¬
    cols = ["í‰ì¼ì£¼ë§", "í˜¸ì„ ", "ì—­ë²ˆí˜¸", "ì—­ëª…", "êµ¬ë¶„"] + \
        [col for col in weekend_avg.columns if col not in ["í‰ì¼ì£¼ë§", "í˜¸ì„ ", "ì—­ë²ˆí˜¸", "ì—­ëª…", "êµ¬ë¶„"]]
    weekend_avg = weekend_avg[cols]

    # í‰ì¼ ë°ì´í„° ì¶”ì¶œ
    weekday = df[df["í‰ì¼ì£¼ë§"] == "í‰ì¼"]

    # ë³‘í•©
    combined = pd.concat([weekday, weekend_avg], ignore_index=True)

    # ì €ì¥ ê²½ë¡œ
    output_path = os.path.join(output_folder, f"cleaned_{filename}")
    try:
        combined.to_csv(output_path, index=False, encoding='euc-kr')
        print(f"âœ… ì €ì¥ ì™„ë£Œ: {output_path}\n")
    except Exception as e:
        print(f"âŒ ì €ì¥ ì‹¤íŒ¨: {e}")




```
