```python
import os
import pandas as pd

# 1. ê²½ë¡œ ì„¤ì •
input_folder = '../2.í˜¼ì¡ë„/2_samename_data'
output_folder = '../2.í˜¼ì¡ë„/3_no_null_data'
os.makedirs(output_folder, exist_ok=True)

# 2. í´ë” ë‚´ ëª¨ë“  íŒŒì¼ ì²˜ë¦¬
for filename in os.listdir(input_folder):
    if not filename.endswith('.csv'):
        continue

    file_path = os.path.join(input_folder, filename)
    print(f"ğŸ“‚ ì²˜ë¦¬ ì¤‘: {filename}")

    try:
        df = pd.read_csv(file_path, encoding='euc-kr')
    except Exception as e:
        print(f"âŒ {filename} ì½ê¸° ì‹¤íŒ¨: {e}")
        continue

    # ì‹œê°„ëŒ€ ì—´ ì¶”ì¶œ (ì•ìª½ ë©”íƒ€ë°ì´í„° ì—´ ì œì™¸)
    time_columns = df.columns[6:]
    df[time_columns] = df[time_columns].apply(pd.to_numeric, errors='coerce')

    # í˜¼ì¡ë„ í•©ì´ 0ì¸ í–‰ ì œê±°
    zero_congestion_mask = (df[time_columns].fillna(0).sum(axis=1) == 0)
    df_cleaned = df[~zero_congestion_mask]

    print(f"ğŸš¨ {filename}: í˜¼ì¡ë„ ì´í•©ì´ 0ì¸ í–‰ {zero_congestion_mask.sum()}ê°œ ì œê±°")

    # ì €ì¥ ê²½ë¡œ
    output_path = os.path.join(output_folder, f"cleaned_{filename}")
    try:
        df_cleaned.to_csv(output_path, index=False, encoding='euc-kr')
        print(f"âœ… ì €ì¥ ì™„ë£Œ: {output_path}\n")
    except Exception as e:
        print(f"âŒ ì €ì¥ ì‹¤íŒ¨: {e}")
```
